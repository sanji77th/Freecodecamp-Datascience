<h3>Modifiying Dataframes</h3>

<h4>Adding Columns</h4>

```python
langs = pd.Series(['French','German','Italian'],
                index = ['France','Germany','Italy'],
                name = 'Language')
```

```python
langs
```

```python
df['Languages'] = langs
```

```python
df
```

<h4>Replace Values</h4>

```python
df['Languages'] = 'English'
```

```python
df
```

<h4>Rename Columns</h4>

```python
df.rename(columns={
    'HDI':'Human Development Index'
},
         index = {
    'United Kingdom':'UK',
    'United States':'USA'
         })
```

```python
df.rename(index=str.upper)
```

```python
df.rename(columns=lambda x:x.lower(),index=str.upper)
```
