```python
import numpy as np
import pandas as pd
```

```python
artists = pd.read_json('Data/artists_nested.json')
```

```python
artists.head()
```

```python
import json

with open('Data/artists_nested.json') as file:
    artsist_dict = json.load(file)
```

```python
artsist_dict
```

```python
from pandas.io.json import json_normalize

bio = json_normalize(artsist_dict,
                    sep='_',
                    record_path='bio',
                    meta=['name'])
```

```python
bio.head()
```

```python
bio.to_json('Data/artist_bio.json')
```
